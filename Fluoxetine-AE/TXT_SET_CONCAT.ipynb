### **DRUG_tmp3_Taming**

```python
import pandas as pd
DRUG_tmp3 = pd.read_csv('/content/drive/MyDrive/RP_A/DRUG_tmp3.csv', sep = ',', on_bad_lines='skip')
DRUG_tmp3.dur_cod.unique()
```

```python
# Drug Usage Duration Mapping
import numpy as np

# Convert columns 'dur' & 'dur_cod' to numeric type
DRUG_tmp3['dur'] = pd.to_numeric(DRUG_tmp3['dur'], errors='coerce')

# Normalize 'dur_cod' so that its values match the keys in our dictionary
DRUG_tmp3['dur_cod'] = DRUG_tmp3['dur_cod'].astype(str).str.upper().str.strip()
```

```python
# Build a dictionary for unit conversion and mapping
age_dict = {
    'DAY': 1,
    'DAYS': 1,
    'YR': 365,
    '    YR': 365,
    '  YR': 365,
    'YEAR': 365,
    'YEARS': 365,
    'YRS': 365,
    'Y': 365,
    'MON': 30,
    'MONTH': 30,
    'MO': 30,
    'MOTH': 30,
    'WK': 7,
    'WKS': 7,
    'WEEKS': 7,
    'MIN': 1/1440,    # Converts minutes to days (1 day = 1440 minutes)
    'HR': 1/24,       # Converts hours to days (1 day = 24 hours)
    'HRS': 1/24,
    '  HRS': 1/24,
    '    HRS': 1/24,
    'HOUR': 1/24,
    'SEC': 1/86400,   # Converts seconds to days (1 day = 86400 seconds)
    'DA': 1,          # Assumes 'DA' stands for days
    'D': 1,
    'CYC': np.nan,    # These codes are mapped to NaN because they are not relevant or not defined
    'TOT': np.nan,
    'TRI': np.nan,
    '2': np.nan,
    'MG': np.nan,
    'UNK': np.nan,
    'AN': np.nan,
}

# Map the normalized duration code to a conversion factor (days)
conversion_factors = DRUG_tmp3['dur_cod'].map(age_dict)

# Calculate dur_days. If either 'dur' or conversion factor is NaN, the result will be NaN.
DRUG_tmp3['dur_days'] = DRUG_tmp3['dur'] * conversion_factors
```

```python
DRUG_tmp3.columns
```

```python
DRUG_tmp3.dur_cod.value_counts()
```

```python
# Group Duration into categories
# Define bins to categorize duration in days:
bins = [0, 30, 60, 90, 120, 150, 180, 360, 100000, np.inf]
labels = ['0-30 d','31-60 d','61-90 d', '91-120 d', '121-150 d', '151-180 d', '181-360 d', '360 d+', '--']
# To handle negative duration days
DRUG_tmp3.loc[DRUG_tmp3['dur_days'] < 0, 'dur_group'] = np.nan
# Categorize 'dur_days' into groups based on bins amd assign to corresponding labels
DRUG_tmp3['dur_group'] = pd.cut(DRUG_tmp3['dur_days'], bins = bins, labels = labels, include_lowest = True)

DRUG_tmp3.loc[DRUG_tmp3['dur_group'] == '--', 'dur_group'] = np.nan
```

```python
# Age Mapping and Conversion
# Convert columns 'age' to numeric type
print(DEMO_tmp.age.dtype)
DEMO_tmp['age'] = pd.to_numeric(DEMO_tmp['age'], errors='coerce')

# Create a dictionary that maps each age unit to the corresponding number of days.
age_dict = {
    'YR': 365,       # 1 Year = 365 days
    'MON': 30,       # 1 Month ≈ 30 days
    'DEC': 3650,     # 1 Decade = 10 years (10 x 365 days)
    'DY': 1,         # 1 Day = 1 day
    'WK': 7          # 1 Week = 7 days
}


# Create a new column 'age_days' by converting the 'age' values into days.
# The conversion factor is determined by mapping the 'age_cod' column using the age_dict.
DEMO_tmp['age_days'] = DEMO_tmp['age'] * DEMO_tmp['age_cod'].map(age_dict)
```

```python
DEMO_tmp
```

```python
# Define bins for categorizing age (in years) and the corresponding labels for each group.
bins = [0, 2, 10, 18, 30, 45, 60, 80, 120, np.inf]
labels = ['0-2','2-10','10-18', '18-30', '30-45', '45-60', '60-80', '80+', '--']

# For any rows where the calculated age in days is negative, set the 'age_group' to NaN.
DEMO_tmp.loc[DEMO_tmp['age_days'] < 0, 'age_group'] = np.nan


# Convert age in days to age in years by dividing by 365, and then use pandas.cut() to assign each value to an age group.
DEMO_tmp['age_group'] = pd.cut(DEMO_tmp['age_days'] / 365, bins=bins, labels=labels, include_lowest=True)


# Remove the original columns 'age', 'age_cod', and 'age_days' as they are no longer needed.
DEMO_tmp.drop(['age_cod', 'age_days'], axis=1, inplace=True)
```

```python
# Define a function to convert different weight scale into 'KG'
# Only consider valid weight units: 'KG', 'LBS' & 'GMS'
def convert_to_kg(df):
  wt = df['wt']
  wt_cod = df['wt_cod']
  try:
    if wt_cod == 'KG':
      return float(wt)
    elif wt_cod == 'LBS':
      return float(wt * 0.453592)
    elif wt_cod == 'GMS':
      return float(wt * 0.001)
    else:
      return np.nan
  except (ValueError, TypeError):
    return np.nan

# Apply the function to create a new column 'weight_kg'
DEMO_tmp['weight_kg'] = DEMO_tmp.apply(convert_to_kg, axis=1)
```

```python
DEMO_tmp.weight_kg.value_counts()
```

```python
# Create a new 'weight_group' column based on proper and reasonable bins and labels
# Define bins for weight in kilograms. These bins are examples and can be adjusted based on your data:
bins = [0, 45, 60, 75, 90, 120, np.inf]
labels = ['0-45 kg','45-60 kg','60-75 kg', '75-90 kg', '90+ kg', '--']
DEMO_tmp.loc[DEMO_tmp['weight_kg'] < 0, 'weight_group'] = np.nan
DEMO_tmp['weight_group'] = pd.cut(DEMO_tmp['weight_kg'], bins=bins, labels=labels, include_lowest=True)

# Remove unnecessary columns.
DEMO_tmp.drop(['wt', 'wt_cod'], axis=1, inplace=True)
```
